```python
import pandas as pd
import matplotlib.pyplot as plt
# Instala gdown para descargar archivos de Google Drive
!pip install gdown

# Obtén el enlace para compartir de tu archivo y reemplaza 'YOUR_FILE_ID'
file_id = '1XHmylmQS9P3NJ-71GviyRaNH2R6WQ8jq'
file_path = f'https://drive.google.com/uc?id={file_id}'

# Descarga el archivo
!gdown {file_path}

!ls


# Reemplaza 'nombre_real_del_archivo.csv' con el nombre real del archivo descargado
datos_descargados = 'datos_descargados.csv'  # Replace with actual file name
df = pd.read_csv(datos_descargados)  # Use the variable 'nombre_del_archivo' here

# Print the column names to verify if 'death_event' exists and check for typos
print(df.columns)

# Calcular las cantidades
anemicos_count = df['anaemia'].sum()
diabeticos_count = df['diabetes'].sum()
fumadores_count = df['smoking'].sum()

# Correct the column name if needed. If the column is named differently,
# replace 'correct_column_name' with the actual name.
muertos_count = df['DEATH_EVENT'].sum()  # Replace 'DEATH_EVENT' with the actual column name if it's different

# Preparar los datos y etiquetas
labels = ['Anaemia', 'Diametes', 'Smoking', 'Death_event']
sizes = [anemicos_count, diabeticos_count, fumadores_count, muertos_count]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Colores opcionales, puedes cambiarlos

# Crear la figura de subplots
plt.figure(figsize=(10, 8))

# Crear cada gráfico de torta en un subplot
for i in range(len(labels)):
    plt.subplot(2, 2, i+1)
    # Adjust explode to match the number of slices in each pie chart
    explode = (0.1, 0)  # Explode the first slice
    plt.pie([sizes[i], sum(sizes)-sizes[i]], labels=[labels[i], 'Otros'], colors=[colors[i], '#cccccc'],
            explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title(f'Distribución de {labels[i]}')

plt.tight_layout()
plt.show()
```
